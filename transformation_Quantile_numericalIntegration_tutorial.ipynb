```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

N = 50000

# MC prediction
R9_MC = np.random.beta(8,3,N)

# Data
R9_Data = np.random.beta(12,2.5,N)
```

```python
bins=np.linspace(0,1,60)

plt.hist(R9_MC,
         bins=bins,
         density=True,
         alpha=0.5,
         label="MC")

plt.hist(R9_Data,
         bins=bins,
         density=True,
         alpha=0.5,
         label="Data")

plt.xlabel("R9")
plt.ylabel("Density")

plt.legend()

plt.show()
```
